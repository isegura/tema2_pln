```python
# ============================================================
# TextCNN multikernel (2,3,4,5) para clasificación de texto
# Google Colab / TensorFlow 2.19
# ============================================================

# ----------------------------
# IMPORTS
# ----------------------------
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import (
    Input, Embedding, Conv1D, BatchNormalization,
    GlobalMaxPooling1D, Concatenate, Dense, Dropout
)
from tensorflow.keras.models import Model

# Google Colab Drive (opcional)
try:
    from google.colab import drive
except ImportError:
    drive = None


# ----------------------------
# CONFIGURACIÓN GLOBAL
# ----------------------------
tf.random.set_seed(42)
print("TensorFlow version:", tf.__version__)


# ----------------------------
# ENTORNO Y RUTAS
# ----------------------------
def get_notebook_environment():
    if "COLAB_GPU" in os.environ:
        return "Google Colab"
    elif "KAGGLE_URL_BASE" in os.environ:
        return "Kaggle"
    return "Entorno local u otro"

entorno = get_notebook_environment()
print("Entorno:", entorno)

if entorno == "Google Colab":
    if drive is not None:
        drive.mount("/content/drive")
    os.chdir("/content/drive/My Drive/Colab Notebooks/data/exist2021/")
elif entorno == "Kaggle":
    os.chdir("/kaggle/input/dataset-exist2021/")

print("Directorio actual:", os.getcwd())


# ----------------------------
# CARGA DE DATOS
# ----------------------------
df_train = pd.read_csv("train.tsv", sep="\t")
df_test = pd.read_csv("test.tsv", sep="\t")

df_train = df_train[df_train["language"] == "en"].copy()
df_test = df_test[df_test["language"] == "en"].copy()

cols_to_drop = ["test_case", "id", "source", "task2"]
df_train.drop(cols_to_drop, axis=1, inplace=True)
df_test.drop(cols_to_drop, axis=1, inplace=True)

df_train.rename(columns={"task1": "label"}, inplace=True)
df_test.rename(columns={"task1": "label"}, inplace=True)

print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)


# ----------------------------
# ETIQUETAS
# ----------------------------
y_train = df_train["label"].tolist()
y_test = df_test["label"].tolist()

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

LABELS = le.classes_
print("Labels:", LABELS)


# ----------------------------
# TOKENIZACIÓN Y PADDING
# ----------------------------
X_train = df_train["text"].tolist()
X_test = df_test["text"].tolist()

tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

NUM_WORDS = len(tokenizer.word_index) + 1
MAX_LEN = 60

X_train_pad = pad_sequences(
    tokenizer.texts_to_sequences(X_train),
    maxlen=MAX_LEN, padding="post", truncating="post"
)
X_test_pad = pad_sequences(
    tokenizer.texts_to_sequences(X_test),
    maxlen=MAX_LEN, padding="post", truncating="post"
)

print("Vocab size:", NUM_WORDS)


# ----------------------------
# MODELO CNN MULTIKERNEL
# ----------------------------
EMBEDDING_SIZE = 100
FILTERS = 128
DROPOUT = 0.4

inputs = Input(shape=(MAX_LEN,), name="input_tokens")

# Embedding: tokens → vectores densos
x = Embedding(NUM_WORDS, EMBEDDING_SIZE, input_length=MAX_LEN, name="embedding")(inputs)

# ------------------------------------------------------------
# Rama kernel = 2 (bigramas)
# ------------------------------------------------------------
c2 = Conv1D(FILTERS, 2, activation="relu", name="conv_k2")(x)

# BatchNormalization:
# - Normaliza las activaciones
# - Reduce variaciones internas
# - Estabiliza y acelera el entrenamiento
c2 = BatchNormalization(name="bn_k2")(c2)

# GlobalMaxPooling1D:
# - Se queda con la activación más fuerte de cada filtro
# - Responde a: “¿aparece este patrón en algún lugar del texto?”
# - Convierte la secuencia en un vector fijo
p2 = GlobalMaxPooling1D(name="gmp_k2")(c2)

# ------------------------------------------------------------
# Rama kernel = 3 (trigramas)
# ------------------------------------------------------------
c3 = Conv1D(FILTERS, 3, activation="relu", name="conv_k3")(x)
c3 = BatchNormalization(name="bn_k3")(c3)
p3 = GlobalMaxPooling1D(name="gmp_k3")(c3)

# ------------------------------------------------------------
# Rama kernel = 4 (4-gramas)
# ------------------------------------------------------------
c4 = Conv1D(FILTERS, 4, activation="relu", name="conv_k4")(x)
c4 = BatchNormalization(name="bn_k4")(c4)
p4 = GlobalMaxPooling1D(name="gmp_k4")(c4)

# ------------------------------------------------------------
# Rama kernel = 5 (5-gramas)
# ------------------------------------------------------------
c5 = Conv1D(FILTERS, 5, activation="relu", name="conv_k5")(x)
c5 = BatchNormalization(name="bn_k5")(c5)
p5 = GlobalMaxPooling1D(name="gmp_k5")(c5)

# Concatenación de todas las features
merged = Concatenate(name="concat_features")([p2, p3, p4, p5])

# Clasificador
h = Dropout(DROPOUT, name="dropout_1")(merged)
h = Dense(128, activation="relu", name="dense_128")(h)
h = Dropout(DROPOUT, name="dropout_2")(h)

outputs = Dense(1, activation="sigmoid", name="output")(h)

model = Model(inputs, outputs, name="TextCNN_MultiKernel")
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()


# ----------------------------
# ENTRENAMIENTO
# ----------------------------
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train_pad,
    y_train,
    epochs=10,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stopping]
)


# ----------------------------
# CURVAS DE ENTRENAMIENTO
# ----------------------------
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="val")
plt.title("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label="train")
plt.plot(history.history["val_loss"], label="val")
plt.title("Loss")
plt.legend()

plt.show()


# ----------------------------
# EVALUACIÓN
# ----------------------------
test_loss, test_acc = model.evaluate(X_test_pad, y_test, verbose=0)
print(f"Test loss: {test_loss:.4f} | Test accuracy: {test_acc:.4f}")

y_prob = model.predict(X_test_pad).ravel()
y_pred = (y_prob >= 0.5).astype(int)

print("Accuracy (sklearn):", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=LABELS))

ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap=plt.cm.Blues)
plt.show()
```
